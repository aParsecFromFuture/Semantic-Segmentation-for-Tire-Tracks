### Global Variables

```python
CLASSES = ['NONE', 'Audi S6', 'BMW X5', 'Chrysler Pacifica', 'Hyundai ix35', 'Mazda 6', 'Renault Megane', 'Toyota Corolla']
SRC_FOLDER = '/content/gdrive/MyDrive/Colab Notebooks/tasarÄ±m/dataset'
```

### Mask Functions

```python
def json_to_mask(json_path):
  with open(json_path, 'r', encoding='utf-8') as f:
    dj = json.load(f)
    
    mask = np.zeros((dj['imageHeight'], dj['imageWidth']), dtype=np.uint8)
    mask = PIL.Image.fromarray(mask)
    draw = PIL.ImageDraw.Draw(mask)

    for shape in dj['shapes']:
      label, points = shape['label'], shape['points']
      idx = CLASSES.index(label) if label in CLASSES else 0
      xy = [tuple(point) for point in points]

      assert len(xy) > 2, 'Polygon must have points more than 2'
      draw.polygon(xy=xy, outline=idx, fill=idx)
    
    return mask
```

### Save

```python
for filepath in glob.glob(f'{SRC_FOLDER}/*/*.json'):
  mask = json_to_mask(filepath)
  mask.save(filepath.replace('.json', '.png'))
```
